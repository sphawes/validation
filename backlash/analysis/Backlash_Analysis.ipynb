```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

mm_per_pixel = 0.08
```

```python
#loading in the csv and converting to a dataframe
path = "../log/xy_precision_1622064954.csv"
data = pd.read_csv(path, sep=r'\s*,\s*', header=0, encoding='ascii', engine='python')

#adding columns for cycle number, and position in mm
data['cycle_number'] = np.arange(len(data))+1
data['x_pos_mm'] = data.apply(lambda row: row.x_pos_pixel * mm_per_pixel, axis=1)
data['y_pos_mm'] = data.apply(lambda row: row.y_pos_pixel * mm_per_pixel, axis=1)

#assigning columns to variables for graphing
x = data['cycle_number']
y1 = data['x_pos_mm']
y2 = data['y_pos_mm']

data.head()
```

```python
fig, axs = plt.subplots(2,1)
axs[0].plot(x, y1)
axs[0].set_xlabel('Cycle Number')
axs[0].set_ylabel('X Position (mm)')

axs[1].plot(x, y2)
axs[1].set_xlabel('Cycle Number')
axs[1].set_ylabel('Y Position (mm)')

fig.tight_layout()
plt.show()
```
